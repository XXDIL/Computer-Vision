### Importing the required Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Utility Functions

```python
def formula_1(B, G, R):
    gray = (B + G + R) // 3
    gray = gray.astype(np.uint8)
    
    return gray
    
def formula_2(B, G, R):
    gray = 0.216*R + 0.7152*G + 0.0722*B
    
    return gray

    
def formula_3(B, G, R):
    gray = 0.299*R + 0.587*G + 0.114*B
    
    return gray
```

### Loading the image and splitting into RGB channels

```python
!wget "https://raw.githubusercontent.com/XXDIL/Computer-Vision/main/images/cat.png"

img = cv2.imread('cat.png')

B = img[:,:,0]
G = img[:,:,1]
R = img[:,:,2]
```

### Results on different methods of Grayscale conversion

- (formula 1) **(R + G + B) / 3**<br/>
- (formula 2) **0.216\*R + 0.7152\*G + 0.0722\*B**<br/>
- (formula 3) **0.299\*R + 0.587\*G + 0.114\*B**

```python
# Processing the images
gray0 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray1 = formula_1(B, G, R)
gray2 = formula_2(B, G, R)
gray3 = formula_3(B, G, R)


# Plotting the images
plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Color Image')

plt.subplot(2,3,2)
plt.imshow(gray0, cmap='gray')
plt.title('Grayscale using OpenCV inbuilt')

plt.subplot(2,3,4)
plt.imshow(gray1, cmap = 'gray')
plt.title('Grayscale using formula 1')

plt.subplot(2,3,5)
plt.imshow(gray2, cmap = 'gray')
plt.title('Grayscale using formula 2')

plt.subplot(2,3,6)
plt.imshow(gray3, cmap = 'gray')
plt.title('Grayscale using formula 3')
           
plt.show()
```
